```python
import folium
from folium.plugins import HeatMap
mapa = folium.Map(location=[imoveis_valor_alto['latitude'].mean(), imoveis_valor_alto['longitude'].mean()], zoom_start=5, tiles='OpenStreetMap')
coordenadas = list(zip(imoveis_valor_alto['latitude'], imoveis_valor_alto['longitude']))

HeatMap(coordenadas).add_to(mapa)

mapa
```

O que faz esses imoveis terem esse valor ?

1 avaliar localização 
2 população das localização
3 renda mediana de quem vive neste locais
4 observa imoveis que estão no mar
5 neste imoveis obeserva total de quartos	total de quartos para dormir	populacao	agregados familiares min e max 



```python
# 1 Avaliar localização 
from sklearn.cluster import KMeans
import numpy as np

# Converter as coordenadas para um formato aceito pelo KMeans
X = np.array(coordenadas)

# Usar KMeans para identificar os clusters
kmeans = KMeans(n_clusters=5)  # Definir o número de clusters
clusters = kmeans.fit_predict(X)

# Adicionar marcadores para os centros dos clusters
centros_clusters = kmeans.cluster_centers_
for centro in centros_clusters:
    folium.Marker(location=[centro[0], centro[1]], popup="Cluster").add_to(mapa)

# Mostrar o mapa
mapa
```

```python
from collections import Counter

# Calcular a contagem de imóveis em cada cluster
contagem_clusters = Counter(clusters)

# Imprimir a contagem de imóveis em cada cluster
for cluster, contagem in contagem_clusters.items():
    print(f"Cluster {cluster + 1}: {contagem} imóveis")
```
